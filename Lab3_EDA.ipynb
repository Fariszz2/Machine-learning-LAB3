```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv("house_prices.csv")

df.head()
```

```python
print(df.isna())
```

```python
print(df.isna().sum())
```

```python
df.duplicated()[df.duplicated()==True]
```

```python
print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python
df.dtypes
```

```python
df['Price (in rupees)'] = df['Price (in rupees)'].replace(r'[$,]', '', regex=True)
df['Price (in rupees)'] = pd.to_numeric(df['Price (in rupees)'], errors='coerce')

df.dtypes
```

```python
df.describe(include='all')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['Price (in rupees)'], bins=20)
plt.title("Distribution of Price (in rupees)")
plt.show()
```

```python
location_price = df.groupby('location')['Price (in rupees)'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
location_price.plot(kind='bar')
plt.title("Total Property Value by Location")
plt.ylabel("Total Price (in rupees)")
plt.show()

location_price
```

```python
location_price = df.groupby('location')['Price (in rupees)'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
location_price.head(10).plot(kind='bar')
plt.title("Total Property Value by Location")
plt.ylabel("Total Price (in rupees)")
plt.xticks(rotation=45)
plt.show()

location_price.head(10)
```

```python
furnishing_avg = df.groupby('Furnishing')['Price (in rupees)'].mean()

furnishing_avg.plot(kind='bar')
plt.title("Average Price by Furnishing Type")
plt.ylabel("Average Price")
plt.show()

furnishing_avg
```

```python
ownership_avg = df.groupby('Ownership')['Price (in rupees)'].mean()

ownership_avg.plot(kind='bar')
plt.title("Average Price by Ownership")
plt.ylabel("Average Price")
plt.show()

ownership_avg
```

```python
df['Bathroom'] = pd.to_numeric(df['Bathroom'], errors='coerce')
clean_df = df[['Bathroom', 'Price (in rupees)']].dropna()

plt.figure(figsize=(8,5))
sns.scatterplot(x='Bathroom', y='Price (in rupees)', data=clean_df)
plt.title("Bathrooms vs House Price")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Price (in rupees)', 'Bathroom']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
